```
class WorkFlow:

    def __init__(self):
        pass

    def task1(self):
        """
            TASK 1:
            ¿Cuál es la distribución de los animes según su género?
        """
        data = Task1Extractor().extract()
        print("Task1 data extracted.")
        data_transformed = Task1Transformer().transform(data)
        print("Task1 data transformed.")
        data_transformed.show()
        Task1Loader(data_transformed).sink()
        print("Task1 data loaded.")
    
    def task2(self):
        """
            TASK 2:
            ¿Cuál es la correlación entre la puntuación (score) y el número de episodios?
        """
        data = Task2Extractor().extract()
        print("Task2 data extracted.")
        data_transformed = Task2Transformer().transform(data)
        print("Task2 data transformed.")
        data_transformed.show()
        Task2Loader(data_transformed).sink()
        print("Task2 data loaded.")


    def task3(self):
        """
            TASK 3:
            ¿Cómo varía la popularidad de los animes a lo largo de los años?
        """
        data = Task3Extractor().extract()
        print("Task3 data extracted.")
        data_transformed = Task3Transformer().transform(data)
        print("Task3 data transformed.")
        data_transformed.show()
        Task3Loader(data_transformed).sink()
        print("Task3 data loaded.")
    
    def task4(self):
        """
            TASK 4:
            ¿Qué géneros de anime son más comúnmente completados por los usuarios?
        """
        data = Task4Extractor().extract()
        print("Task4 data extracted.")
        data_transformed = Task4Transformer().transform(data)
        print("Task4 data transformed.")
        data_transformed.show()
        Task4Loader(data_transformed).sink()
        print("Task4 data loaded.")

    def runner(self):  
        self.task1()
        self.task2()
        self.task3()
        self.task4()



```

```
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("de_project_mal").getOrCreate()

workflow = WorkFlow().runner()
```
